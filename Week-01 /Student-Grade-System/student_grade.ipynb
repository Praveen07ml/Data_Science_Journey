## Tasks

- Create a dictionary to store student name and their marks
- Make Functions 1. store new student and their marks
                 2. calculate class average
                 3. Find Top Student
                 4. students above average
                 5. Display All studnets
- Menu Card

```python
student = {}

def add_student(name,marks):
     student[name] = marks

add_student('praveen',89)
add_student('dhanush',81)
add_student('veeru',68)
add_student('kaarchi',45)
print(student)
print()

def average():
    if len(student) == 0:
        return 0
    total = 0

    for key,value in student.items():
        total = total + value
    return total / len(student)

print(average())
print()

def top_student():
    maximum = 0
    name = " "
    for key,value in student.items():
        if value > maximum:
            maximum = value
            name = key
    return name,maximum

print(top_student())
print()

def above_avg():
    avg = average()
    above_avg = {}
    for key,value in student.items():
        if value > avg:
            above_avg[key] = value
    return above_avg

print(above_avg())
print()

def show_student():
    for key,value in student.items():
        print(key,"-",value)

show_student()


# menu where you can make your choice 

while True:

    print("\n 1.Add Student")
    print("2.Show Student")
    print("3.Class Average")
    print("4.Top Student")
    print("5.Student Above Average")
    print("6.Exit\n")

    choice = input("Enter your choice : ")
    print()

    if choice == "1":
        name = input('Enter name :')
        marks = int(input('Enter marks'))
        add_student(name,marks)
        print("Students info added Successfully")
        
    elif choice == "2":
        show_student()
        
    elif choice == "3":
        print("Class Average :",average())
        
    elif choice == "4":
        print("Top Studnet :",top_student())

    elif choice == "5":
        print("Students Above average :",above_avg())

    elif choice == "6":
        print("Exit the program")
        break

    else:
        print("Invalid Choice")
        
        
        
```
